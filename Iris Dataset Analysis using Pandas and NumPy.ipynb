```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
# Load Iris dataset
iris = load_iris()

# Create DataFrame
df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

# Rename columns
df.columns = [
    'sepal_length',
    'sepal_width',
    'petal_length',
    'petal_width'
]

# Add species column
df['species'] = pd.Categorical.from_codes(
    iris.target,
    iris.target_names
)

# Display first few rows
df.head()
```

```python
print("First 10 Rows:")
df.head(10)
```

```python
print("Unique Species:")
print(df['species'].unique())
```

```python
print("Flower Count by Species:")
species_count = df['species'].value_counts()

print(species_count)
```

```python
df['species'] = df['species'].astype(str)

avg_petal_length = df.groupby(
    'species'
)['petal_length'].mean()

print(avg_petal_length)
```

```python
print("Flower with Maximum Sepal Length:")

max_sepal_flower = df.loc[
    df['sepal_length'].idxmax()
]

print(max_sepal_flower)
```

```python
sorted_df = df.sort_values(
    by='petal_width',
    ascending=False
)

print("Sorted by Petal Width (Descending):")

sorted_df.head(10)
```

```python
filtered_df = df[
    df['petal_length'] > 5
]

print("Flowers with Petal Length > 5")

filtered_df.head()
```

```python
df['total_size'] = (
    df['sepal_length']
    + df['petal_length']
)

print(df.head())
```

```python
filtered_df.to_csv(
    "filtered_iris.csv",
    index=False
)

print("Filtered data saved as filtered_iris.csv")
```

```python
petal_array = df['petal_length'].to_numpy()

print("NumPy Array:")

print(petal_array)
```

```python
mean_value = np.mean(petal_array)
median_value = np.median(petal_array)
std_value = np.std(petal_array)

print("Mean =", mean_value)
print("Median =", median_value)
print("Standard Deviation =", std_value)
```

```python
max_value = np.max(petal_array)
min_value = np.min(petal_array)

print("Maximum Petal Length =", max_value)
print("Minimum Petal Length =", min_value)
```

```python
normalized_petal = (
    petal_array - np.min(petal_array)
) / (
    np.max(petal_array) - np.min(petal_array)
)

print("Normalized Petal Lengths:")

print(normalized_petal)
```

```python
print("First 10 Normalized Values:")

print(normalized_petal[:10])
```
