```python
import pandas as pd
```

```python
net=pd.read_csv("/content/netflix_advanced_dataset.csv")
net
```

```python
net['type'].value_counts()
```

```python
print("Total number of movies")
movies=net[net["type"]=="Movie"]
len(movies)
```

```python
movies=net[(net["type"]=="Movie") & (net["release_year"]>2020)]
print("Movie released after 2020")
movies
```

```python
oldest_movie=net[(net["type"]=="Movie")&(net["release_year"].min())]
print("Oldest movie is ")
oldest_movie
```

```python
old_movie=net[net["release_year"]==net["release_year"].min()]
print("Oldest movie is ")
old_movie
```

```python
print("number of oldest movies")
len(old_movie)
```

```python
country_count=net['country'].value_counts()
print("Counting movies by country ")
country_count
```
